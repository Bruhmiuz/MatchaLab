```python
# IMPLEMENT FRACTION
# DO NOT CHANGE THIS CELL

def gcd(a: int, b: int):
    if not a:
        if not b:
            return 0
        return b
    if not b:
        return a

    if abs(b) > abs(a):
        a, b = abs(b), abs(a)
    else:
        a, b = abs(a), abs(b)

    while a != 1:
        if b % a == 0:
            break

        a, b = b % a, a

    return a

class fraction:
    def __init__(self, num: int = 0, denom: int = 1):
        if isinstance(denom, int):
            num *= denom // abs(denom)
            denom = abs(denom)
        elif isinstance(denom, fraction):
            num *= denom / abs(denom.num)
            num *= denom.denom
            denom = abs(denom.num)
        else:
            assert ValueError("fraction() only accepts int or fraction input")

        if isinstance(num, fraction):
            denom *= num.denom
            num = num.num
        elif not isinstance(num, int):
            assert ValueError("fraction() only accepts int or fraction input")

        cd = gcd(num, denom)
        self.num = num // cd
        self.denom = denom //cd


    def __neg__(self):
        return fraction(-self.num, self.denom)


    def __abs__(self):
        return fraction(abs(self.num), self.denom)


    def __add__(self, other):
        if isinstance(other, fraction):
            num = self.num * other.denom + other.num * self.denom
            denom = self.denom * other.denom
            return fraction(num, denom)

        elif isinstance(other, int):
            return fraction(self.num + other * self.denom, self.denom)

        elif isinstance(other, float):
            return (self.num / self.denom) + other

        else:
            return type(other).__radd__(other, self)

    def __radd__(self, other):
        return self + other


    def __sub__(self, other):
        if isinstance(other, fraction):
            num = self.num * other.denom - other.num * self.denom
            denom = self.denom * other.denom
            return fraction(num, denom)

        elif isinstance(other, int):
            return fraction(self.num - other * self.denom, self.denom)

        elif isinstance(other, float):
            return (self.num / self.denom) - other

        else:
            return type(other).__rsub__(other, self)

    def __rsub__(self, other):
        return -self + other


    def __truediv__(self, other):
        if isinstance(other, fraction):
            num = self.num * other.denom
            denom = self.denom * other.num
            return fraction(num, denom)

        elif isinstance(other, int):
            return fraction(self.num, other * self.denom)

        elif isinstance(other, float):
            return (self.num / self.denom) / other

        else:
            return type(other).__rtruediv__(other, self)

    def __rtruediv__(self, other):
        return fraction(1, 1) / (self / other)


    def __mul__(self, other):
        if isinstance(other, fraction):
            num = self.num * other.num
            denom = self.denom * other.denom
            return fraction(num, denom)

        elif isinstance(other, int):
            return fraction(other * self.num, self.denom)

        elif isinstance(other, float):
            return (self.num / self.denom) * other

        else:
            return type(other).__rmul__(other, self)

    def __rmul__(self, other):
        return self * other


    @property
    def val(self):
        return self.num / self.denom

    def __eq__(self, other):
        return self.val == other

    def __gt__(self, other):
        return self.val > other

    def __lt__(self, other):
        return self.val < other

    def __ge__(self, other):
        return self.val >= other

    def __le__(self, other):
        return self.val <= other

    def __str__(self):
        return f"{self.val:.3f}"

    def __repr__(self):
        return f"fraction({self.num}, {self.denom})"
```

```python
# IMPLEMENT VECTOR
# DO NOT CHANGE THIS CELL

# Similar to list but supports vector and scalar operations
# For arithmetic operations, just one side needs to be a vector, the other can be a list

class vector(list):
    # TBD: implement dimensionality

    def __neg__(self):
      new_vec = vector(self)
      for i in range(len(self)):
        new_vec[i] = -new_vec[i]
      return new_vec


    def __matmul__(self, other):
        if isinstance(other, list):
            if len(self) != len(other):
                raise IndexError(f"Inconsistent vector size: {len(self)} and {len(other)}")
            mat = vector([vector([0] * len(self)) for _ in range(len(self))])
            for i, v1 in enumerate(self):
                for j, v2 in enumerate(other):
                    mat[i][j] = v1 * v2
            return mat

        else:
            return type(other).__rmatmul__(other, self)

    def __rmatmul__(self, other):
        if isinstance(other, list):
            return vector(other) @ self


    def __mul__(self, other): # scalar multiplication
        if isinstance(other, list): # element wise multiplication
            if len(self) != len(other):
                raise IndexError(f"Inconsistent vector size: {len(self)} and {len(other)}")
            new_vec = vector(self)
            for i in range(len(self)):
                new_vec[i] *= other[i]

        else:
            new_vec = vector(self)
            for i in range(len(self)):
                new_vec[i] *= other

        return new_vec

    def __rmul__(self, other):
        return self * other


    def __truediv__(self, other):
        if isinstance(other, list):
            if len(self) != len(other):
                raise IndexError(f"Inconsistent vector size: {len(self)} and {len(other)}")
            new_vec = vector(self)
            for i in range(len(self)):
                new_vec[i] /= other[i]

        else:
            new_vec = vector(self)
            for i in range(len(self)):
                new_vec[i] /= other

        return new_vec

    def __rtruediv__(self, other):
        return other / self


    def __add__(self, other):
        if isinstance(other, list):
            if len(self) != len(other):
                raise IndexError(f"Inconsistent vector size: {len(self)} and {len(other)}")
            new_vec = vector(self)
            for i, v in enumerate(other):
                new_vec[i] += v

        else:
            new_vec = vector(self)
            for i in range(len(self)):
                new_vec[i] += other

        return new_vec

    def __radd__(self, other):
        return self + other


    def __sub__(self, other):
        if isinstance(other, list):
            if len(self) != len(other):
                raise IndexError(f"Inconsistent vector size: {len(self)} and {len(other)}")
            new_vec = vector(self)
            for i, v in enumerate(other):
                new_vec[i] -= v

        else:
            new_vec = vector(self)
            for i in range(len(self)):
                new_vec[i] -= other

        return new_vec

    def __rsub__(self, other):
        return -self + other


    def to_fraction(self):
        for i in range(len(self)):
            self[i] = fraction(self[i])
        return self


    def __str__(self):
        mxl = 0
        cpy = [0] * len(self)
        for i, v in enumerate(self):
            cpy[i] = str(v)
            mxl = max(mxl, len(cpy[i]))

        vec_str = "["
        for item in cpy:
            vec_str += " " * (mxl - mxl // 2 - (len(item) - len(item) // 2) + 1) + item + " " * (mxl // 2 - len(item) // 2 + 1)
        vec_str += "]"

        return vec_str


def dot_product(a: list, b: list):
    if len(a) != len(b):
        raise IndexError
    res = 0
    for i in range(len(a)):
        res += a[i] * b[i]
    return res
```

```python
# IMPLEMENT MATRIX
# DO NOT CHANGE THIS CELL
"""
Methods:

to_fraction - VERY IMPORTANT!! Normalizes all INT/FRACTION values within the matrix into FRACTION (makes computing rational numbers easier), does not work on FLOAT so do it beforehand!!!
              Changes the current matrix and additionally returns MATRIX

rows - Returns the matrix as a 2D VECTOR
cols - Returns the transposed matrix as a 2D VECTOR
__matmul__ (matrixA @ matrixB) - Performs matrix multiplication in the order of AB. Returns MATRIX

=============================================================================================================================================
Notice: To access the i-th row/col in the following methods, please input i-1 (for example, the 1st row would be matrix[0,:] instead of matrix[1,:])

getitem (mat[INT/SLICE, INT/SLICE]) - Returns the selected cells as either a VALUE, a 1D VECTOR or 2D VECTOR depending on the input
setitem (mat[INT/SLICE, INT/SLICE] = val) - Accepts a VALUE, a 1D VECTOR or 2D VECTOR (make sure the dimensions are correct!!!), and sets the selected cells as the input accordingly.
row_swap(a,b) - Swaps rows a and b. Changes the current matrix and additionally returns MATRIX
col_swap(a,b) - Swaps cols a and b. Changes the current matrix and additionally returns MATRIX

=============================================================================================================================================
Notice: The following methods perform insertion on the current matrix, up to size + 1 different positions can be chosen
        pos = 0 performs insertion before the 1st line, pos = i performs insertion after the i-th line
        pos = -1 performs insertion after the last line, pos = -i performs insertion before the i-th last line

expand_rows(pos, other: 2D LIST/2D VECTOR/MATRIX) - Inserts rows, at the end by default. Changes the current matrix and additionally returns MATRIX
expand_cols(pos, other: 2D LIST/2D VECTOR/MATRIX) - Inserts cols, at the end by default. Changes the current matrix and additionally returns MATRIX
augment_line(pos) - Inserts the augment line, before the last line by default. If it exists after the last line, it is not shown during print.
                    The augment line prevents further operations from being done on the right hand side of the augmented matrix.
                    Changes the current matrix and additionally returns MATRIX

=============================================================================================================================================

ref(show_steps: BOOL = false, find_lower: INT = 0) - Performs gaussian elimination algorithm to reduce the current matrix into row-echelon form. Note that the algorithm may not resemble gaussian elimination done by hand,
                                                     since the algorithm does not have vision and I'm also tring to avoid using sort (generally not a good practice)
                                                     Does not affect current matrix.
                                                     find_lower = 0 will return the reduced (U) matrix as a MATRIX (default)
                                                     find_lower = 1 will return the L and U matrices (or PL and U matrices) as MATRIX, MATRIX
                                                     find_lower = 2 will return the P, L and U matrices as MATRIX, MATRIX, MATRIX

rref(show_steps: BOOL = false) - Calls self.ref() first and performs back substitution to reduce the current matrix into reduced row-echlon form. Steps shown may not be efficient as a result.
                                 Does not affect current matrix. Returns the matrix in rref form as a MATRIX

identity() - Returns the identity matrix (I) based on row size (for reasons) as a MATRIX

lower() - Returns the lower triangular (L or PL) matrix as a MATRIX

inverse() - Returns the inverse matrix as a MATRIX
            (If determinant is zero, returns a pseudo inverse matrix)

To be implemented:
scalar addition
scalar multiplication
determinant
"""

class Matrix:
    def __init__(self, rows, augment: int = None):
        if isinstance(rows, Matrix): # copy
            self._size = rows._size
            self._items = rows._items[:]
            self._augment = rows._augment
        else:
            size = (len(rows), len(rows[0]))
            for row in rows[1:]:
                if len(row) != size[1]:
                    raise IndexError("Inconsistent matrix size")
            self._size = size # resize to be implemented if needed
            self._items = []
            self._augment = augment if augment else self._size[1]
            for item in rows:
                self._items += list(item)

    @property
    def rows(self): # to 2d vector
        rows = vector([vector([0] * self._size[1]) for _ in range(self._size[0])])
        for r in range(self._size[0]):
            for c in range(self._size[1]):
                rows[r][c] = self._items[r * self._size[1] + c]
        return rows

    @property
    def cols(self): # to 2d vector
        cols = vector([vector([0] * self._size[0]) for _ in range(self._size[1])])
        for c in range(self._size[1]):
            for r in range(self._size[0]):
                cols[c][r] = self._items[r * self._size[1] + c]
        return cols


    def select_idx(self, idx: slice, size):
        if idx.start is None:
            start = 0
        elif idx.start < 0:
            start = max(0, idx.start + size)
        else:
            start = min(size, idx.start)

        if idx.stop is None:
            stop = size
        elif idx.stop < 0:
            stop = max(0, idx.stop + size)
        else:
            stop = min(size, idx.stop)

        return slice(start, stop, 1)

    def __getitem__(self, key: tuple):
        if isinstance(key[0], slice):
            sel_r = self.select_idx(key[0], self._size[0])
            if isinstance(key[1], slice): # return smaller matrix (2d vector)
                sel_c = self.select_idx(key[1], self._size[1])
                mat = vector([vector([0] * (sel_c.stop - sel_c.start)) for _ in range(sel_r.stop - sel_r.start)]) # initiate smaller matrix
                for i, r in enumerate(range(sel_r.start, sel_r.stop, sel_r.step)):
                    for j, c in enumerate(range(sel_c.start, sel_c.stop, sel_c.step)):
                        mat[i][j] = self._items[r * self._size[1] + c]
                return mat

            elif isinstance(key[1], int): #return smol col (vector)
                c = key[1] + self._size[1] if key[1] < 0 else key[1]
                col = vector([0] * (sel_r.stop - sel_r.start))
                for i, r in enumerate(range(sel_r.start, sel_r.stop, sel_r.step)):
                    col[i] = self._items[r * self._size[1] + c]
                return col

        elif isinstance(key[0], int):
            r = key[0] + self._size[0] if key[0] < 0 else key[0]
            if isinstance(key[1], slice): # return smol row (vector)
                sel_c = self.select_idx(key[1], self._size[1])
                row = vector([0] * (sel_c.stop - sel_c.start))
                for j, c in enumerate(range(sel_c.start, sel_c.stop, sel_c.step)):
                    row[j] = self._items[r * self._size[1] + c]
                return row

            elif isinstance(key[1], int): # return item
                c = key[1] + self._size[1] if key[1] < 0 else key[1]
                # print(key[0], key[1])
                return self._items[r * self._size[1] + c]

    def __setitem__(self, key: tuple, value):
        if isinstance(key[0], slice):
            sel_r = self.select_idx(key[0], self._size[0])

            if isinstance(key[1], slice):
                sel_c = self.select_idx(key[1], self._size[1])

                for i, r in enumerate(range(sel_r.start, sel_r.stop, sel_r.step)):
                    for j, c in enumerate(range(sel_c.start, sel_c.stop, sel_c.step)):
                        self._items[r * self._size[1] + c] = value[i][j]

            elif isinstance(key[1], int):
                c = key[1] + self._size[1] if key[1] < 0 else key[1]
                for i, r in enumerate(range(sel_r.start, sel_r.stop, sel_r.step)):
                    self._items[r * self._size[1] + c] = value[i]

        elif isinstance(key[0], int):
            r = key[0] + self._size[0] if key[0] < 0 else key[0]
            if isinstance(key[1], slice):
                sel_c = self.select_idx(key[1], self._size[1])

                for j, c in enumerate(range(sel_c.start, sel_c.stop, sel_c.step)):
                    self._items[r * self._size[1] + c] = value[j]

            elif isinstance(key[1], int):
                c = key[1] + self._size[1] if key[1] < 0 else key[1]
                self._items[r * self._size[1] + c] = value

    def __mul__(self, other):
        if isinstance(other, (int, float, fraction)):
            new_mat = Matrix(self)
            for i in range(len(self._items)):
                new_mat._items[i] = self._items[i] * other
            return new_mat

    def __rmul__(self, other):
        if isinstance(other, (int, float)):
            return self * other

    def __matmul__(self, other):
        if isinstance(other, list):
            if self._size[1] != len(other):
                raise IndexError(f"Inconsistent matrix and vector size: {self._size[1]} and {len(other)}")
            new_mat = Matrix([[0] for _ in range(self._size[0])])
            for r in range(self._size[0]):
                new_mat._items[r] = dot_product(self[r,:], other)
            return new_mat

        elif isinstance(other, Matrix):
            if self._size[1] != other._size[0]:
                raise IndexError(f"Inconsistent matrix size: {self._size[1]} and {other._size[0]}")
            new_mat = Matrix([[0] * other._size[1] for _ in range(self._size[0])])
            for r in range(self._size[0]):
                for c in range(other._size[1]):
                    new_mat._items[r * other._size[1] + c] = dot_product(self[r,:], other[:,c])
            return new_mat

    def __rmatmul__(self, other):
        if isinstance(other, list):
            if self._size[0] != len(other):
                raise IndexError(f"Inconsistent matrix and vector size: {len(other)} and {self._size[0]}")
            new_mat = Matrix([[0] * self._size[1]])
            for c in range(self._size[1]):
                new_mat._items[c] = dot_product(other, self[:,c])
            return new_mat

    def __str__(self):
        mxl = 0
        syscpy = [[0] * self._size[1] for _ in range(self._size[0])]
        for r in range(self._size[0]):
            for c in range(self._size[1]):
                syscpy[r][c] = str(self._items[r * self._size[1] + c])
                mxl = max(mxl, len(syscpy[r][c]))

        mat_str = ""
        for row in syscpy:
            mat_str += "["
            for i, item in enumerate(row):
                if i == self._augment:
                    mat_str += " | "
                mat_str += " " * (mxl - mxl // 2 - (len(item) - len(item) // 2) + 1) + item + " " * (mxl // 2 - len(item) // 2 + 1)
            mat_str += "]\n"

        return mat_str

    def __repr__(self):
        mat_repr = f"Matrix(["
        for r in range(self._size[0]):
            mat_repr += f"{self._items[r * self._size[1]:(r + 1) * self._size[1]]}"
            if r < self._size[0] - 1:
                mat_repr += ", "
        if self._augment < self._size[1]:
            mat_repr += f"], {self._augment})"
        else:
            mat_repr += "])"

        return mat_repr

    def row_swap(self, a: int, b: int): # eg. matrix.row_swap(0,1) swaps the first 2 rows
        self[a, :], self[b, :] = self[b, :], self[a, :]
        return self

    def col_swap(self, a: int, b: int):
        self[:, a], self[:, b] = self[:, b], self[:, a]
        return self

    def to_fraction(self):
        for i in range(len(self._items)):
            self._items[i] = fraction(self._items[i])
        return self


    # adds more rows to the bottom by default
    #if pos >= 0, starts from before the first row (pos = 0) to after xth row (pos = x)
    #else if pos < 0, starts from after the last row (pos = -1) to before xth last row (pos = -1-x)

    def expand_rows(self, other, pos: int = -1):
        pos = max(0, min((pos + self._size[0] + 1 if pos < 0 else pos), self._size[0]))

        if isinstance(other, list):
            if len(other) != self._size[1]:
                raise IndexError("erm, what the sigma")
            up = self._items[:pos * self._size[1]]
            down = self._items[pos * self._size[1]:]
            self._items = up + list(other) + down # convert to list cuz we don't want it to do vector addition
            self._size = (self._size[0] + 1, self._size[1])

        elif isinstance(other, Matrix):
            if other._size[1] != self._size[1]:
                raise IndexError("erm, what the sigma")
            up = self._items[:pos * self._size[1]]
            down = self._items[pos * self._size[1]:]
            self._items = up + other._items + down
            self._size = (self._size[0] + other._size[0], self._size[1])

        return self


    # same as defined in expand_rows except for columns
    def expand_cols(self, other, pos: int = -1):
        pos = max(0, min((pos + self._size[1] + 1 if pos < 0 else pos), self._size[1]))

        if isinstance(other, list):
            if len(other) != self._size[0]:
                raise IndexError("erm, what the sigma")
            temp = self._items
            self._size = (self._size[0], self._size[1] + 1)
            self._items = [0] * (self._size[0] * self._size[1])
            for r in range(self._size[0]):
                for left in range(pos):
                    self._items[r * self._size[1] + left] = temp[r * (self._size[1] - 1) + left]

                self._items[r * self._size[1] + pos] = other[r]
                for right in range(pos + 1, self._size[1]):
                    self._items[r * self._size[1] + right] = temp[r * (self._size[1] - 1) + right - 1]

            if self._augment >= pos:
                self._augment += 1

        elif isinstance(other, Matrix):
            if other._size[0] != self._size[0]:
                raise IndexError("erm, what the sigma")
            temp = self._items
            self._size = (self._size[0], self._size[1] + other._size[1])
            self._items = [0] * (self._size[0] * self._size[1])
            for r in range(self._size[0]):
                for left in range(pos):
                    self._items[r * self._size[1] + left] = temp[r * (self._size[1] - other._size[1]) + left]

                for c in range(other._size[1]):
                    self._items[r * self._size[1] + pos + c] = other._items[r * other._size[1] + c]

                for right in range(pos + other._size[1], self._size[1]):
                    self._items[r * self._size[1] + right] = temp[r * (self._size[1] - other._size[1]) + right - other._size[1]]

            if self._augment >= pos:
                self._augment += other._size[1]

        return self


    def identity(self): # creates identity matrix based on ROW SIZE ONLY (self._size[0])
        temp = [[0] * self._size[0] for _ in range(self._size[0])]
        for i in range(self._size[0]):
            temp[i][i] = 1

        return Matrix(temp)

    def transpose(self):
        return Matrix(self.cols)

    def augment_line(self, pos: int = -2): #
        self._augment = max(0, min((pos + self._size[1] + 1 if pos < 0 else pos), self._size[1]))
        return self

    @property
    def coefficients(self):
        return self[:, :self._augment]

    @property
    def constants(self):
        return self[:, self._augment:]

    def ref(self, show_steps: bool = False, nonpivot_cols: list = None, find_lower = 0):
        new_mat = Matrix(self)
        a_rows = self._size[0]
        a_cols = self._augment
        total_cols = self._size[1]
        zero_row_count = 0
        nonpivot_col_count = 0
        x = 0

        if find_lower:
            lower = self.identity()
            perm = Matrix(lower)

        while x < a_cols:
            if x - nonpivot_col_count >= a_rows - zero_row_count:
                nonpivot_col_count += 1
                if nonpivot_cols is not None:
                    nonpivot_cols.append(x)
                x += 1
                continue

            curr_zero_count = 0
            while new_mat[x - nonpivot_col_count, x] == 0:
                curr_zero_count += 1
                if x >= a_rows - (zero_row_count + curr_zero_count):
                    nonpivot_col_count += 1
                    if nonpivot_cols is not None:
                        nonpivot_cols.append(x)
                    break

                if new_mat[-(zero_row_count + curr_zero_count), x] != 0:
                    new_mat.row_swap(x - nonpivot_col_count, -(zero_row_count + curr_zero_count))              # E: Swap row
                    if show_steps:
                        print(f"R{x-nonpivot_col_count+1} <-> R{a_rows-(zero_row_count+curr_zero_count)+1}")
                        print(new_mat)

                    if find_lower:
                        perm.row_swap(x - nonpivot_col_count, -(zero_row_count + curr_zero_count))

            else:
                R, C = x - nonpivot_col_count, x
                for r in range(R + 1, a_rows):
                    if r >= a_rows - zero_row_count:
                        break

                    zero_row = True
                    while zero_row:
                        mul = new_mat[r, C] / new_mat[R, C]
                        for c in range(C, a_cols): # subtraction is done element-wise instead of whole row to keep track of zeros (all reduced to zero = zero row)
                            new_mat[r, c] -= new_mat[R, c] * mul
                            if new_mat[r, c] != 0:
                                zero_row = False

                        new_mat[r, a_cols:] -= new_mat[R, a_cols:] * mul

                        if mul != 0:
                            if show_steps:                                                                     # E: Add/sub row
                                curr_step = f"R{r+1}"
                                if mul > 0:
                                    curr_step += " - "
                                else:
                                    curr_step += " + "
                                if isinstance(mul, fraction):
                                    if mul.denom == 1:
                                        if mul.num == 1:
                                            curr_step += f"R{R+1}"
                                        else:
                                            curr_step += f"{abs(mul.num)}R{R+1}"
                                    else:
                                        curr_step += f"{abs(mul.num)}/{mul.denom}R{R+1}"
                                else:
                                    curr_step += f"{abs(mul)}R{R+1}"
                                print(curr_step)
                                print(new_mat)

                            if find_lower:
                                lower[r, C] += mul

                        if zero_row:
                            zero_row_count += 1
                            if r == a_rows - zero_row_count:
                                break

                            new_mat.row_swap(r, -zero_row_count)
                            if show_steps:                                                                     # E: Swap row
                                print(f"R{r+1} <-> R{a_rows-zero_row_count+1}")
                                print(new_mat)

                            if find_lower:
                                perm.row_swap(r, -zero_row_count)

            x += 1

        if find_lower == 1:
            return perm @ lower, new_mat

        elif find_lower == 2:
            return perm, lower, new_mat

        return new_mat



    def rref(self, show_steps: bool = False):
        nonpivot_cols = [] # stores nonpivot cols

        new_mat = self.ref(show_steps, nonpivot_cols)

        a_rows = self._size[0]
        a_cols = self._augment
        total_cols = self._size[1]

        # (len(nonpivot_cols) + self._size[0] - self._augment) is the amount of nonzero cols
        i = a_rows - (len(nonpivot_cols) + self._size[0] - self._augment) - 1
        for j in range(a_cols - 1, -1, -1):
            if nonpivot_cols and j == nonpivot_cols[-1]:
                nonpivot_cols.pop()
                continue

            if new_mat[i, j] != 1:
                mul = 1 / new_mat[i, j]
                new_mat[i, j:] *= mul

                if show_steps:
                    if isinstance(mul, fraction):
                        if mul.denom == 1:
                            if mul.num == -1:
                                print(f"-R{i+1}")
                            else:
                                print(f"{mul.num}R{i+1}")
                        else:
                            print(f"{mul.num}/{mul.denom}R{i+1}")
                    else:
                        print(f"{mul}R{i+1}")
                    print(new_mat)

            for r in range(i - 1, -1, -1):
                if new_mat[r, j] == 0:
                    continue

                mul = new_mat[r, j]
                if mul != 0:
                    new_mat[r, j:] -= vector(new_mat[i, j:]) * mul
                    if show_steps:
                        curr_step = f"R{r+1}"
                        if mul > 0:
                            curr_step += " - "
                        else:
                            curr_step += " + "
                        if isinstance(mul, fraction):
                            if mul.denom == 1:
                                if mul.num == 1:
                                    curr_step += f"R{i+1}"
                                else:
                                    curr_step += f"{abs(mul.num)}R{i+1}"
                            else:
                                curr_step += f"{abs(mul.num)}/{mul.denom}R{i+1}"
                        else:
                            curr_step += f"{abs(mul)}R{i+1}"
                        print(curr_step)
                        print(new_mat)

            i -= 1

        return new_mat

    def inverse(self):
        if self._size[0] != self._size[1]:
            raise IndexError
        # if self.det == 0:
        #     raise ValueError
        return Matrix(Matrix(self).expand_cols(self.identity()).rref()[:, self._size[1]:]) #yummers


    def lower(self):
        return self.ref(find_lower = 1)[0]


def create_diag(size, n: int = 1):
    temp = [[0] * size for _ in range(size)]
    for i in range(size):
        temp[i][i] = n
    return Matrix(temp)

def sum_2d(mat):
    if isinstance(mat, list):
        res = 0
        for row in mat:
            for item in row:
                res += item
        return res

    elif isinstance(mat, Matrix):
        return sum(mat._items)

    else:
        raise TypeError
```
